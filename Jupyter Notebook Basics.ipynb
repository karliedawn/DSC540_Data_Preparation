```python
import math

def tuple_to_int(stack):
    r_number = 0
    for i in range(0,len(stack)):
        r_number = r_number + (stack.pop()*math.pow(10,i))
    return r_number
```

5. Reach each line and pass that line to a function, along with the list of headers. The work of the function is to construct a dict out of these two and fill up the **key:values**. Keep in mind that a missing value should result in **None**.

```python
# This function takes the header of the csv file and the current line 
# and adds it to the dictionary

def list_to_dict(columnnames, linelist):
    sales_dict = {}
    for n in range(len(columnnames)):
        sales_dict[columnnames[n]] = linelist[n]
    print(sales_dict)
```

```python
# this function turns csv file into dictionaries

with open('sales_record.csv','r') as fd:
    first_line = fd.readline()
    column_names = first_line.replace('\n','').split(',')
    for i, line in enumerate(fd):
        line_list = line.replace('\n','').split(',')
        list_to_dict(column_names, line_list)
        if i>3:
            break
```
