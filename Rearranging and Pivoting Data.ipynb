```python
# Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import with proper encoding, dropping most of the extraneous quotes, skipping lines that don't have enough column values
met_df = pd.read_csv('MetObjects.csv', encoding = 'iso-8859-1', quoting=3, on_bad_lines='skip', index_col=False, skipinitialspace=True)
```

```python
# Look at dataframe
met_df
```

### Step 1: Drop Missing Data

```python
# Drop columns with more than 50% NA values
met_df.dropna(thresh=5624, inplace=True, axis=1)
```

```python
met_df.shape
```

```python
# Drop rows with missing Object ID
met_df = met_df[met_df['Object ID'].notna()]
```

```python
met_df.shape
```

### Step 2: Remove Duplicates

```python
# How many duplicates
met_df.duplicated().sum()
```

```python
# Drop duplicates
met_df = met_df.drop_duplicates()
```

```python
met_df.shape
```

### Step 3: Create hierarchical index

```python
# Indexing with dataframe's columns
index_df = met_df.set_index(['Title', 'Object ID'])
```

```python
index_df
```

### Step 5: Grouping with Functions

### Step 6: Split/Apply/Combine

```python
# returns object number
met_df.groupby(['Object Name', 'Object End Date']).apply(lambda x: x['"Object Number'])
```

### Step 7: Convert between string and date time

```python
# Convert these columns to datetimes
met_df['Object Begin Date'] = pd.to_datetime(met_df['Object Begin Date'], errors = 'coerce')
met_df['Object End Date'] = pd.to_datetime(met_df['Object End Date'], errors = 'coerce')
```

```python
#dates subset
dates_df = met_df[['Object Begin Date','Object End Date']]
```

```python
dates_df.dropna(inplace=True)
```
